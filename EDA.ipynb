```python
#is all the movies have sub-parts --> no
#some movies have the parts so fo those moves we have collection name
#so where null value is there i replace it with "No collection"
#before doing above steps i want to convert the string form of dictionary into particular collections
#to convert into particular collection we use a module --> ast
```

```python
import ast
import numpy as np
#Store the values in new list
Fa = []
for i in df['belongs_to_collection']:
    if i == i:
      Fa.append(ast.literal_eval(i)['name'])
    else:
      Fa.append("No Collection")
```

```python
#Create A New Columns Callled Collection
df['Collection_name'] = Fa
```

```python
df.head()
```

```python
#We Name New Collection Columns So Drop The Old One.
df = df.drop(columns=['belongs_to_collection'])
```
