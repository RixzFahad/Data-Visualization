**Question-3** *What Are Natural Keys?*

* Primary Key:
  * A primary key is a column or set of columns that uniquely identifies each record in a table.

* Natural Key:
  * A natural key is a real-world attribute that already exists in the data and can uniquely identify a record.
  * It is derived from business data and has meaningful information.

* Examples of Natural Keys:
  * Aadhaar Number
  * PAN Card Number
  * Mobile Number
  * Email ID
  * Passport Number

* Characteristics of Natural Keys:
  * Comes from real business data
  * Has meaning in the real world
  * Usually already present in the dataset
  * Can sometimes change (which is a drawback)

* Surrogate Key:
  * A surrogate key is an artificial or system-generated key used to uniquely identify records.
  * It has no business meaning and is usually created by the database.

* Examples of Surrogate Keys:
  * Customer_ID
  * Order_ID
  * Product_ID
  * Auto-increment numbers

* Characteristics of Surrogate Keys:
  * Generated by the system
  * Has no business meaning
  * Stable and does not change
  * Widely used in data warehouses and fact tables

* Star Schema and Snowflake Schema

  * These schemas help structure data for efficient querying and reporting
  * In both schemas, the central table is the Fact Table
  * To build relationships between tables, we follow these schema designs
  * By default, Power BI uses Star Schema which is a "denormalized design"
  * Snowflake Schema is a "normalized design", meaning tables are split into multiple sub-parts
