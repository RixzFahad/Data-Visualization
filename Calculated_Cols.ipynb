**Q) Why does the `total_price` column show the same price in every row?**

**Answer**

*Because DAX calculates the **SUM of the entire column** for every row when it is used inside a **Calculated Column***.

---

**Explanation**

**DAX works with two important contexts**

* **Row Context**
  - Calculation happens **row by row**
  - Each row is evaluated individually

* **Filter Context**
  - Calculation changes depending on:
    - visuals
    - slicers
    - filters
    - grouping

---

**What happens in a Calculated Column**

* DAX evaluates the formula for **each row**
* But **`SUM()` calculates the total of the entire column**
* There is **no filter context**
* Therefore the **same total value appears in every row**

---

**Example**

| Product | Price |
|--------|------|
| A | 100 |
| B | 200 |
| C | 300 |

**SUM(Price) = 600**

---

**Result**

| Product | Price | Total_Price |
|--------|------|-------------|
| A | 100 | 600 |
| B | 200 | 600 |
| C | 300 | 600 |
