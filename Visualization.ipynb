***Box Plot***

*Is Used To Detect The Outliers In The Data*

*   Box Plot USed To Check the Distribution Of Continous Data W.R.T Categorical Data.



```python
import seaborn as sns
#Syntax:-  sns.boxplot(data =, x=, y=,)
sns.boxplot(data=df, x='sex', y='total_bill',hue='day')
plt.show()

# what are all component do we have in the box?
# 1. 5% minimum  --> lower limit
# 2. Q1 (quantile)  --> 25% percentile of data
# 3. Q2  --> 50% percentile of the data
# 4. Q3  --> 75% percentile of the data
# 5. 95% maximum  --> 95% --> upper limit
```
