*Power BI / DAX – Count Functions*

- **COUNT()**
  - *Counts only numeric values in a column.*

  - **Syntax**
    COUNT(Column_Name)

  - **Example: Sales Table**

    Product | Quantity  
    A       | 10  
    B       | 5  
    C       | (blank)  
    D       | 7  
    E       | Text  

  - **DAX**
    Numeric_Count = COUNT(Sales[Quantity])

  - **Result**
    3  
    *(Counts only 10, 5, and 7)*


- **COUNTA()**
  - *Counts all non-blank values (numbers + text).*

  - **Syntax**
    COUNTA(Column_Name)

  - **Example: Sales Table**

    Product | Quantity  
    A       | 10  
    B       | 5  
    C       | (blank)  
    D       | 7  
    E       | Text  

  - **DAX**
    NonBlank_Count = COUNTA(Sales[Quantity])

  - **Result**
    4  
    *(Counts 10, 5, 7, and Text)*


- **Summary**

  - **COUNT()** → *Counts only numeric values*  
  - **COUNTA()** → *Counts all non-blank values*
