***6. IFERROR***

→ *IFERROR function* returns a **specific value** if the expression results in an error  
→ Otherwise, it returns the **actual expression result**

***Syntax***

IFERROR(value, <value_if_error)

***Example (DAX)***

Result =
IFERROR(
    'New SalesTransactions'[Sales] / 'New SalesTransactions'[Quantity],
    0
)

***Explanation***

→ If division causes error (like divide by 0) → returns **0**  
→ Otherwise → returns actual calculated value

***IFERROR Explanation***

→ When we perform **division**, dividing by **0** may cause an **error**  
→ To avoid this, we use **IFERROR**

→ If there is any **error in a cell** and we want to apply a **condition** in that column,  
use **IFERROR**

***Example***

Safe Division =
IFERROR(
    'New SalesTransactions'[Sales] / 'New SalesTransactions'[Quantity],
    "Error Handled"
)
