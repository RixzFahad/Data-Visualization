#  **Data Hierarchy (Power BI)**

  
  * **Definition:** Data hierarchy allows users to drill down from higher-level data to lower-level detailed data inside reports or visuals.

* **Example Hierarchy Structure**  
    * **Year → Quarter → Month → Day**

* **How it Works**  
    * **Year:** Shows overall yearly summary  
    * **Quarter:** Breaks the year into 4 parts  
    * **Month:** Shows monthly performance  
    * **Day:** Provides the most detailed level of analysis

* **Purpose of Data Hierarchy**  
    * **Drill-down Analysis:** Move from summary to detailed insights  
    * **Better Data Navigation:** Easily explore different levels of data  
    * **Interactive Reports:** Users can expand and collapse levels in visuals

* **Definition in Simple Words**  
    * **Data hierarchy is a structured arrangement of related fields that allows users to navigate through different levels of data from summary to detailed level.**

* **How to Build a Hierarchy in Power BI**  
    * **Step 1:** Open **Model View** or **Data View**  
    * **Step 2:** In the **Fields Pane**, right-click on the main column  
    * **Step 3:** Click **Create Hierarchy**  
    * **Step 4:** A new hierarchy will be created  
    * **Step 5:** Drag other related columns and **Add to Hierarchy**

* **Example: Date Hierarchy**  
    * **Year**  
    * **Quarter**  
    * **Month**  
    * **Day**

* **Other Common Hierarchies**  
    * **Country → State → City**  
    * **Category → Subcategory → Product**  
    * **Department → Team → Employee**

* **Data Category (Power BI)**  
    * **Definition:** Data Category tells Power BI what type of data a column contains so Power BI can use it correctly in visuals like maps, URLs, or geographic analysis.

* **Purpose of Data Category**  
    * **Improve Data Recognition:** Helps Power BI understand the meaning of the column.  
    * **Better Visualizations:** Enables features like maps and web links.  
    * **Accurate Analysis:** Ensures data is interpreted correctly.

* **How to Set Data Category in Power BI**  
    * **Step 1:** Go to **Table View**  
    * **Step 2:** **Select the field (column)** you want to categorize  
    * **Step 3:** Look for **Data Category** in the ribbon/menu  
    * **Step 4:** **Select the appropriate option** based on your data type

* **Common Data Category Examples**  
    * **Country/Region** → Used for map visuals  
    * **State or Province** → Geographic location analysis  
    * **City** → Map-based reporting  
    * **Postal Code** → Location identification  
    * **Web URL** → Clickable links in reports  
    * **Image URL** → Display images inside Power BI tables

# **Best Practices of Data Modelling (Power BI)**



* **Load Required Tables**
    * **Load only the specific tables or queries into Model View**
    * Avoid loading unnecessary data to keep the model clean.

* **Hide Unnecessary Fields**
    * **Hide unwanted fields from tables**
    * This makes reports easier to use and reduces confusion.

* **Identify Unique Keys**
    * **Find unique keys (Primary Keys) in tables**
    * These keys help create proper **relationships between tables**.

* **Use Star Schema (Most Important)**
    * **Build relationships between Fact and Dimension tables**
    * Fact table → contains **numeric values (sales, revenue, quantity)**
    * Dimension table → contains **descriptive data (date, product, customer)**

* **Create Data Hierarchy**
    * **Create hierarchies from tables** for drill-down analysis
    * Example: **Year → Quarter → Month → Day**

* **Use Data Categorization**
    * **Assign Data Category to fields**
    * Helps visualize **maps, links, and images correctly**

* **M Language (Power BI)**  
    * **Meaning:** M Language is a data transformation language used in **Power Query (Power BI Back End)**.

* **Purpose of M Language**
